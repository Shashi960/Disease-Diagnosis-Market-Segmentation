```python
encode_cols = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting', 'Like','Gender']
```

```python
# Yes = 1 , No = 0 and Male =1 Female = 0

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in encode_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
print(df['yummy'].value_counts())
print(df['convenient'].value_counts())
print(df['VisitFrequency'].value_counts())
```

```python
from sklearn.preprocessing import OneHotEncoder
en = OneHotEncoder()

# fit and transform the dataframe column
df['VisitFrequency'] = en.fit_transform(df[['VisitFrequency']]).toarray()

df.head()
```

```python
sns.pairplot(df)
```

```python

# plot histogram of ages
plt.figure(figsize=(10, 10))
sns.histplot(x='Age', data=df)
plt.title('Age Distribution')
plt.show()


```

```python
# plot scatter plot of age vs. like
plt.figure(figsize=(10, 10))
sns.scatterplot(x='Age', y='Like', data=df)
plt.title('Age vs. Like')
plt.show()

```

```python
# plot bar plot of visit frequency
plt.figure(figsize=(10, 10))
sns.countplot(x='VisitFrequency', data=df)
plt.title('Visit Frequency')
plt.show()


```

```python
# plot heatmap of correlation matrix
plt.figure(figsize=(10, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
